```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
app_data = pd.read_csv('../data/apriori_data.csv')
app_list = [app_data.loc[i][0].split(',') for i in range(len(app_data))]
```

```python
# Implement Apriori algorithm
from apyori import apriori
rules = apriori(transactions=app_list,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=3)
```

```python
result_list = list(rules)
```

```python
# Create a function to extract the parameters from the result of apriori algorithm
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0]+','+tuple(result[2][0][1])[1] if len(result[2][0][1]) > 1 else tuple(result[2][0][1])[0] for result in results]
    support = [result[1] for result in results]
    confidence = [result[2][0][2] for result in results]
    lift = [result[2][0][3] for result in results]
    return list(zip(lhs,rhs,support,confidence,lift))
```

```python
# Create a dataframe for a better view
rules_df = pd.DataFrame(inspect(result_list),columns=['Left Hand Side','Right Hand Side', 'Support', 'Confidence', 'Lift'])
rules_df
```
